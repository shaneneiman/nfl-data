```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("datafiles/plays.csv")
```

```python
df.describe()
```

```python
for col in df.columns:
    print(col)
```

```python
# Scatter Plot

ax1 = df[df['passResult'] == 'C'].plot(kind='scatter', x='defendersInTheBox', y='offensePlayResult', color='blue', alpha=0.5, figsize=(10, 7))
df[df['passResult'] == 'I'].plot(kind='scatter', x='defendersInTheBox', y='offensePlayResult', color='magenta', alpha=0.5, figsize=(10 ,7), ax=ax1)
plt.legend(labels=['Complete Pass', 'Incomplete Pass'])
plt.title('Relationship between Complete and Incomplete Passes', size=24)
plt.xlabel('Number of Defenders in the Box', size=18)
plt.ylabel('Play Resut (yards)', size=18);
```

```python
# Scatter Plot

df.plot(kind='scatter', x='yardsToGo', y='offensePlayResult', color='blue', alpha=0.5, figsize=(10, 7))
plt.title('Relationship between Yards to Go and Play Result', size=24)
plt.xlabel('Yards to Go', size=18)
plt.ylabel('Play Resut (yards)', size=18);
```
