```python
# Read CSV Files
plays_df = pd.read_csv("datafiles/plays.csv")
players_df = pd.read_csv("datafiles/players.csv")
week1_df = pd.read_csv("datafiles/week1.csv")
week2_df = pd.read_csv("datafiles/week2.csv")
week3_df = pd.read_csv("datafiles/week3.csv")
```

```python
# join play data with weekly data
week1plays_df = week1_df.join(plays_df, lsuffix="_left")
week2plays_df = week2_df.join(plays_df, lsuffix="_left")
week3plays_df = week3_df.join(plays_df, lsuffix="_left")
```

```python
stats_positive = week1plays_df[week1plays_df["playResult"] > 0].corr()
stats_positive.rename(columns=lambda x: x + " positive gain", inplace=True)

stats_negative = week1plays_df[week1plays_df["playResult"] < 0].corr()
stats_negative.rename(columns=lambda x: x + " negative gain", inplace=True)

stats_zero = week1plays_df[week1plays_df["playResult"] == 0].corr()
stats_zero.rename(columns=lambda x: x + " zero gain", inplace=True)

stats_positive
```
